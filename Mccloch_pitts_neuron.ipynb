```python
#Mccloch-Pitts Neuron AND, OR & XOR.

import matplotlib.pyplot as plt
import time
import numpy as np

inputs = [(0,0),(0,1),(1,0),(1,1)]
threshold_or = 1
threshold_and = 2
weights = [(1,1)]

fig, axes = plt.subplots(1,3,figsize=(15,5))
print("|  x1 | x2  | or_o/p | and_o/p |   xor_o/p   |")
for x1,x2 in inputs:
  for w1,w2 in weights:
    y_or = x1*w1 + x2*w2
    y_and = x1*w1 + x2*w2
    or_output = 1 if y_or >= threshold_or else 0
    and_output = 1 if y_and >= threshold_and else 0
    xor_output = 1 if (x1 + x2) % 2 == 1 else 0

    print(f"|  {x1}  |  {x2}  |   {or_output}    |    {and_output}    |      {xor_output}      |")

    color_or = 'red' if or_output == 1 else 'blue'
    color_and = 'red' if and_output == 1 else 'blue'
    color_xor = 'red' if xor_output == 1 else 'blue'
    axes[0].scatter(x1,x2,color=color_or,s=100)
    axes[1].scatter(x1,x2,color=color_and,s=100)
    axes[2].scatter(x1, x2, color=color_xor, s=100)

x_vals = np.linspace(-0.5, 1.5, 100)
y_or = (threshold_or - x_vals * weights[0][0]) / weights[0][1]
axes[0].plot(x_vals, y_or, 'k--', label="Decision Boundary")

y_and = (threshold_and - x_vals * weights[0][0]) / weights[0][1]
axes[1].plot(x_vals, y_and, 'k--', label="Decision Boundary")
axes[2].plot([], [], 'k--')

axes[0].plot(x1,x2,'o',color=color_or)
axes[1].plot(x1,x2,'o',color=color_and)
axes[2].plot(x1, x2, 'o', color=color_xor)
axes[0].set_title("OR Gate")
axes[1].set_title("AND Gate")
axes[2].set_title("XOR Gate")
axes[0].set_xlabel("x1")
axes[0].set_ylabel("x2")
axes[1].set_xlabel("x1")
axes[1].set_ylabel("x2")
axes[2].set_xlabel("x1")
axes[2].set_ylabel("x2")
axes[0].grid(True)
axes[1].grid(True)
axes[2].grid(True)
axes[0].legend()
axes[1].legend()
time.sleep(1)
plt.show()
print("\n")
```
